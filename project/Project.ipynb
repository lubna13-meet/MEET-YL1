```python
import csv
```

```python
f = open('data.csv', 'rb')
reader = csv.reader(f)
cig = []
leuk = []
for row in reader:
    a = row [0]
    b = a.replace ("\t" ," ")
    c = b.split()
    cig.append(c[1])
    leuk.append(c[5])
cig = map(float,cig)
leuk = map(float,leuk)
print cig
print leuk
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('No. of cigarettes & leukemia victims')
ax.set_xlabel('No. of leukemia victims')
ax.set_ylabel('No. of cigarettes')
ax.scatter(cig, leuk)
```
