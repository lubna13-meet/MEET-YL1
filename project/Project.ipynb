```python
import csv
```

```python
f = open('data.csv', 'rb')
reader = csv.reader(f)
cig = []
leuk = []
for row in reader:
    a = row [0]
    b = a.replace ("\t" ," ")
    c = b.split()
    cig.append(c[1])
    leuk.append(c[5])
cig = map(float,cig)
leuk = map(float,leuk)
print cig
print leuk
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('No. of cigarettes & leukemia victims')
ax.set_ylabel('No. of leukemia victims per 100K person')
ax.set_xlabel('No. of cigarettes')
ax.scatter(cig, leuk)
```

This scatter plot shows the number of cigarettes that are smoked in 44 states and the percentage of people that die because of leukemia which is caused because of cigarettes in the same state.

```python
f = open('data.csv', 'rb')
reader = csv.reader(f)
blad = []
lung = []
for row in reader:
    a = row [0]
    b = a.replace ("\t" ," ")
    c = b.split()
    blad.append(c[2])
    lung.append(c[3])
blad = map(float,blad)
lung = map(float,lung)
print blad
print lung
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Percentage of bladder cancer & lung cancer victims')
ax.set_xlabel('Bladder cancer victims')
ax.set_ylabel('Lung cancer victims')
ax.bar(blad,lung, alpha= 0.3)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Percentage of bladder cancer & lung cancer victims')
ax.set_xlabel('Bladder cancer victims')
ax.set_ylabel('Lung cancer victims')
ax.scatter(blad,lung, alpha= 0.3)
```

These 2 plots (the bare and the scatter one have the same data but i made the 2 plots to show how it looks when it is bars and how it looks when it is in scatter) show both the percantage of people dying because bladder cancer and people dying because of lung cancer in 44 different states. These deaths are caused by smoking cigarettes. so these plots show the percentage of people dying because of different cancers in the same state.
